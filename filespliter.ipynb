```python
import os 

lines_per_file = 50000
smallfile = None
big_file_name = 'data/spikenuc0523.fasta'
small_file_name_format = 'data/spikenuc0523/spikenuc0523_{}.fasta'

split_file_path = 'data/spikenuc0523'

isExist = os.path.exists(split_file_path) 

count = 0
if not isExist:
    os.makedirs(split_file_path)

with open(big_file_name) as bigfile:
    for lineno, line in enumerate(bigfile):
        if lineno % lines_per_file == 0:
            if smallfile:
                smallfile.close()
            small_filename = small_file_name_format.format(int(lineno/lines_per_file +1))
            smallfile = open(small_filename, "w")
            print(' creating {}'.format(small_filename))
        smallfile.write(line)
        count +=1
    if smallfile:
        smallfile.close()
        count +=1
        

print("finished , total {} files".format(count))        
```
